<h2>Visualization</h2>

Investigate the relationship between your chosen response variable and at least two predictors using visualizations. Write down your observations.

**Visualization 1**

```python
df.plot.scatter(x="INSTRUCTION_EXPENDITURE_PER_STUDENT", y="AVG_READING_4_SCORE", alpha = 0.6)
plt.xlabel("Instruction Expenditure Per Student")
plt.ylabel("Average 4th Grade Reading Score")
plt.title("Relationship Between Instruction Expenditure Per Student and Average 4th Grade Reading Score")
plt.show()
```

This scatter plot shows the relationship between the instruction expenditure per student and the average 4th grade reading score. There is a positive relationship between the two variables, meaning that as the instruction expenditure per student increased, the average 4th grade reading score also increased.

**Visualization 2**

```python
df.plot.scatter(x="FEDERAL_REVENUE_PER_STUDENT", y="AVG_READING_4_SCORE", alpha = 0.6)
plt.xlabel("Federal Revenue Per Student")
plt.ylabel("Average 4th Grade Reading Score")
plt.title("Relationship Between Federal Revenue Per Student and Average 4th Grade Reading Score")
plt.show()
```

This scatter plot shows the relationship between the federal revenue per student and the average 4th grade reading score. There is a constant relationship between the two variables at first but then there is a negative relationship, meaning that as the federal revenue per student increased, the average 4th grade reading score initially stayed the same but then started to decrease.

<h2> Data Creation </h2>

_Use this space to create train/test data_

```python
from sklearn.model_selection import train_test_split
```

```python
X = df[['ENROLL', 'AVG_MATH_4_SCORE', 'AVG_READING_8_SCORE']].dropna()
y = df.loc[X.index]['AVG_READING_4_SCORE']
```

```python
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=42)
```

<h2> Prediction </h2>

ML Models [Resource](https://medium.com/@vijaya.beeravalli/comparison-of-machine-learning-classification-models-for-credit-card-default-data-c3cf805c9a5a)

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
```

```python
model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
```

## Evaluation

Choose some metrics to evaluate the performance of your model, some of them are mentioned in the tutorial.

```python
print(np.mean(model.predict(X_test)-y_test))
print(np.mean(np.abs(model.predict(X_test)-y_test)))
```

We have copied over the graphs that visualize the model's performance on the training and testing set. 

Change `col_name` and modify the call to `plt.ylabel()` to isolate how a single predictor affects the model.

```python
col_name = 'AVG_MATH_4_SCORE'

f = plt.figure(figsize=(12,6))
plt.scatter(X_train[col_name], y_train, color = "red")
plt.scatter(X_train[col_name], model.predict(X_train), color = "green")

plt.legend(['True Training','Predicted Training'])
plt.xlabel(col_name)
plt.ylabel('AVG_READING_4_SCORE')
plt.title("Model Behavior On Training Set")
```

```python
col_name = 'AVG_READING_8_SCORE'

f = plt.figure(figsize=(12,6))
plt.scatter(X_test[col_name], y_test, color = "blue")
plt.scatter(X_test[col_name], model.predict(X_test), color = "black")

plt.legend(['True testing','Predicted testing'])
plt.xlabel(col_name)
plt.ylabel('AVG_READING_4_SCORE')
plt.title("Model Behavior on Testing Set")
```
